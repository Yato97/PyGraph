## I) Import des librairies nécessaires 

```python
from PyGraph.Pygraph import *
```

## II) Création d'un graphe non orienté représentant une grille de jeu 10x10 

### Initialisation des sommets 

```python
G = Graph(100, engine="neato")
```

```python
G.view
```

### Initialisation des arêtes 

```python
index = 9
for x in range(0,99, 10):
    for i in range(x, index):
        x = i + 1
        G.add_edge(i,x)
        if x+9 <= 99:
            G.add_edge(i,x+9)
    index += 10
for x in range(9,99, 10):
    i = x + 10
    G.add_edge(i,x)
    
# 'circo', 'dot', 'fdp', 'neato', 'osage', 'patchwork', 'sfdp', 'twopi'
```

```python
G.view
```

## III) Positionnement des sommets & dimension du graphe

```python
x = 0
y = 0
tab_pos = list()
for i in range(0,100, 1):
    if x%10 == 0:
        y+=1
        x = 0
    temp = [i,x,y]
    tab_pos.append(temp)
    x += 1
G.position(tab_pos)
G.scale(0.5)
G.view
```

```python
G.reset_view(engine="patchwork")
G.color_on(0,'red')
G.view
```
