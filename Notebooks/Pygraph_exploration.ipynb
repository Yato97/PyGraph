# Présentation des méthodes concernant le module Pygraph

## I) Import des librairies nécessaires 

```python
from PyGraph.Pygraph import *
from PyGraph.path_finder import *
```

## II) Générer des graphes 

### 1) Graphe non orienté aléatoire ou pas sans pondération

Création du graphe et de ces sommets

```python
G1 = Graph(4)
G1.view
```

Ajouter un lien entre 2 sommets

```python
G1.add_edge(0,1)
G1.view
```

Ajouter des liens aux sommets

```python
G1.add_edges_from([(2,3), (1,2), (2,0), (3,1)])
G1.view
```

Retirer un lien entre 2 sommets

Retirer des liens aux sommets

Aléatoire

```python
GA = Graph(6, random=True, engine='neato')
GA.view
```

### 2) Graphe orienté sans pondération

```python
G2 = DiGraph(5)
G2.add_edges_from([(0,1), (1,0), (1,3), (3,2), (3,4), (2,4), (0,2)])
G2.view
```

### 3) Graphe orienté pondéré

```python
G3 = DiGraph(9, strict=True)
G3.add_edges_from([(0, 1, 3), (0, 3, 6), (1, 2, 2),(1, 5, 20), (1, 4, 6), (2, 5, 4),(3, 4, 4), (5, 7, 3), (4, 7, 22),(6, 7, 19), (6, 8, 21), (7, 8, 18), (3, 6, 7)])
G3.view
```

## III) Changer l'algorithme utilisé pour l'affichage du graphe

```python
G2.reset_view(engine='circo')
G2.view
```

## IV) Positionner les sommets

```python
G3.position([[0, -4, 0], [3, 2, 0], [1, 0, 3], [2, -4, 5], [4, 2, 3], [6, 5, 3], [5, 0, 5], [8, 8, 5], [7, 5, 5]])
G3.view
```

## V) Changer les étiquettes

```python
G3.set_labels("ABCDEFGHI")
G3.label_on()
G3.view
```

## VI) Changer la couleur d'un sommets

```python
G3.color_on(0,'red')
G3.view
```

## VII) Changer la couleur d'un arc

```python
G3.color_on_edge(0,1,'purple')
G3.view
```

## VIII) Dimensionner un graphe et ses sommets

```python
G3.scale(0.5)
G3.resize(0.4)
G3.view
```

## IX) Copie d'un graphe déjà existant 

```python
G4 = G3.copy()
G4.view
```

## X) Sauver sour forme d'image dans un fichier

```python
G4.write(filename='output', format='svg')
```

## XI) Charger & sauvegarder un graphe avec le format JSON

```python
G4.save_json(filename='G4json', encoding='utf-8')
```
